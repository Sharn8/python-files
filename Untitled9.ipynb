```python
import random

choices=["rock","paper","scissors"]

while True:
  user_choice=input("Enter rock, paper, scissors (or 'quit' to exit)")

if user_choice=="quit":
  print("thanks for playing")
  break

if user_choice not in choices:
  print("invalid choice, try again")
  continue

computer_choice=random.choice(choices)

print(f"\n you chose:{user_choice}")
print(f"computer chose:{computer_choice}")

if user_choice==computer_choice:
  print("it's a tie")

elif(user_choice=="rock" and computer_choice=="scissors") or (user_choice=="scissors" and computer_choice=="paper") and (user_choice=="paper" and computer_choice=="rock"):
  print("you win")
else:
  print("you lost")
```

#TWO SUM

```python
def twosum(nums,target):
  hashmap ={}
  for i, num in enumerate(nums):
    complement = target-num

    if complement in hashmap:
         return [hashmap[complement],i]

    hashmap[num]=i
  return[]

print(twosum([2,7,11,15],9))
print(twosum([3,2,4],6))
print(twosum([3,3],6))
```
